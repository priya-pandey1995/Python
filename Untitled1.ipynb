```python
import matplotlib.pyplot as plt
import numpy as np

# Define various activation functions and their descriptions
activation_functions = {
    'Sigmoid': {
        'function': lambda x: 1 / (1 + np.exp(-x)),
        'description': 'The sigmoid function smoothly maps any input to a value between 0 and 1.'
    },
    'ReLU': {
        'function': lambda x: np.maximum(0, x),
        'description': 'ReLU (Rectified Linear Unit) sets all negative values to zero and passes positive values unchanged.'
    },
    'Tanh': {
        'function': lambda x: np.tanh(x),
        'description': 'Tanh squashes input to the range [-1, 1] and is zero-centered.'
    },
    'Leaky ReLU': {
        'function': lambda x: np.where(x > 0, x, 0.01 * x),
        'description': 'Leaky ReLU prevents zero gradients for negative inputs by allowing a small negative slope.'
    }
}

# Generate x values
x = np.linspace(-5, 5, 1000)

# Plotting activation functions
plt.figure(figsize=(12, 10))

for idx, (func_name, func_data) in enumerate(activation_functions.items(), start=1):
    plt.subplot(2, 2, idx)
    y = func_data['function'](x)
    plt.plot(x, y, label=func_name)
    plt.title(f'{func_name} Activation Function')
    plt.xlabel('x')
    plt.ylabel(f'{func_name.lower()}(x)')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()

# Print descriptions
for func_name, func_data in activation_functions.items():
    print(f"{func_name} Function Description:\n{func_data['description']}\n")
```
